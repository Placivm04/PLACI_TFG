Los ataques adversarios son técnicas que manipulan las entradas para provocar errores en modelos de IA.

Es decir:

- No atacan al sistema directamente
- Atacan a los datos que el modelo recibe

El atacante intentaría mover una muestra de Ataque -> Normal sin que parezca un ataque real modificado.

**<h3>Tipos</h3>**

Entre los ataques adversarios tenemos principalmente 2 tipos:

1. Poisoning (en entrenamiento)
2. Evasión (en inferencia) <---- El nuestro

Un ataque de evasión consiste en:

- Coger una muestra maliciosa
- Difícil para detectar por humanos
- Muy efectivos contra modelos complejos

Ejemplo:

Tenemos esto que distinguimos como ataque
Ataque real:
[bytes=5000, conexiones=100, duración=2s]

El atacante genera:
[bytes=4800, conexiones=92, duración=2.1s]

Sigue siendo ataque, pero el modelo lo detecta como normal

**<h3>Generar ataques basados en BÚSQUEDA</h3>**

Tenemos diferentes tipos de ataques tipo:
- Random perturbation
- Hill climbing
- Genetic algorithms
- Boundary attack
